```python
import pandas as pd
#Question-1A
data = {
    'Student_ID': [f'Student_{i+1}' for i in range(20)],
    'Score': [88, 76, 92, 45, 67, 53, 81, 39, 95, 62,
              70, 84, 55, 47, 73, 58, 60, 49, 91, 66]
}

df = pd.DataFrame(data)
average_score = df['Score'].mean()
highest_score = df['Score'].max()
lowest_score = df['Score'].min()

print("=== Overall Class Performance ===")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score: {lowest_score}")
df['Result'] = df['Score'].apply(lambda x: 'Pass' if x >= 50 else 'Fail')

print("\n=== Pass/Fail Classification ===")
print(df[['Student_ID', 'Score', 'Result']])
def assign_grade(score):
    if score >= 90:
      return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'

    elif score >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Score'].apply(assign_grade)

print("\n=== Grading (Predefined Thresholds) ===")
print(df[['Student_ID', 'Score', 'Grade']])
```

```python
#Question 2B
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Manually enter grades for 30 students
# Each list should have 30 grades (Math, Science, English)
math_grades = [85, 78, 90, 67, 88, 76, 95, 82, 60, 70,
               92, 84, 89, 73, 65, 80, 75, 68, 93, 77,
               86, 71, 79, 88, 62, 83, 74, 81, 69, 91]

science_grades = [72, 84, 88, 90, 76, 81, 69, 85, 80, 77,
                  92, 64, 70, 95, 68, 74, 66, 91, 86, 73,
                  87, 79, 63, 82, 88, 75, 93, 78, 71, 89]

english_grades = [90, 82, 85, 75, 79, 88, 70, 65, 83, 77,
                  91, 80, 72, 76, 68, 84, 66, 93, 89, 87,
                  73, 81, 78, 90, 74, 85, 69, 92, 67, 86]

# Step 2: Create DataFrame
grades = pd.DataFrame({
    'Math': math_grades,
    'Science': science_grades,
    'English': english_grades
})

# Step 3: Calculate total and average per student
grades['Total'] = grades.sum(axis=1)
grades['Average'] = grades[['Math', 'Science', 'English']].mean(axis=1)

# Step 4: Show subject-wise statistics
print("=== Subject-wise Statistics ===")
for subject in ['Math', 'Science', 'English']:
    print(f"\n{subject} Grades:")
    print("Average:", grades[subject].mean())
    print("Highest:", grades[subject].max())
    print("Lowest :", grades[subject].min())

# Step 5: Overall stats
print("\n=== Overall Statistics ===")
print("Class Average (All Subjects):", grades[['Math', 'Science', 'English']].values.mean())
print("Highest Total Score:", grades['Total'].max())
print("Lowest Total Score:", grades['Total'].min())

# Step 6: Top 5 students by average
print("\n=== Top 5 Students (Based on Average Score) ===")
print(grades.sort_values(by='Average', ascending=False).head(5))

# Step 7: Optional Plot
plt.figure(figsize=(12, 4))
for i, subject in enumerate(['Math', 'Science', 'English']):
    plt.subplot(1, 3, i + 1)
    plt.hist(grades[subject], bins=8, color='lightgreen', edgecolor='black')
    plt.title(f'{subject} Grade Distribution')
    plt.xlabel('Grade')
    plt.ylabel('Students')
plt.tight_layout()
plt.show()
```

```python
#Question-3A
import numpy as np

# Functions to calculate impedance
def resistor(R):
    return complex(R)

def capacitor(C, freq):
    if C == 0:
        return complex(np.inf)
    return complex(0, -1 / (2 * np.pi * freq * C))

def inductor(L, freq):
    return complex(0, 2 * np.pi * freq * L)

# Series and parallel impedance calculators
def series_impedance(impedances):
    return sum(impedances)

def parallel_impedance(impedances):
    if not impedances:
        return 0
    inverse_total = sum(1 / z for z in impedances)
    return 1 / inverse_total if inverse_total != 0 else complex(np.inf)

# Main program
def main():
    print("Simple Impedance Calculator")
    freq = float(input("Enter frequency in Hz: "))
    total_components = int(input("Enter number of components: "))

    series_list = []
    parallel_list = []

    for i in range(total_components):
        print(f"\nComponent {i+1}")
        ctype = input("Type (R = Resistor, C = Capacitor, L = Inductor): ").upper()
        value = float(input("Value (Ohms, Farads, or Henrys): "))
        connection = input("Connection (S = Series, P = Parallel): ").upper()

        # Calculate impedance
        if ctype == 'R':
            z = resistor(value)
        elif ctype == 'C':
            z = capacitor(value, freq)
        elif ctype == 'L':
            z = inductor(value, freq)
        else:
            print("Invalid component type. Skipping...")
            continue

        # Add to corresponding list
        if connection == 'S':
            series_list.append(z)
        elif connection == 'P':
            parallel_list.append(z)
        else:
            print("Invalid connection type. Skipping...")
            continue

    # Calculate total impedance
    z_series = series_impedance(series_list)
    z_parallel = parallel_impedance(parallel_list)
    total = z_series + z_parallel

    print(f"\nTotal Impedance: {total:.2f} Ω")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}


df = pd.DataFrame(data)


df['Total Value'] = df['Price'] * df['Quantity']


most_expensive_product = df.loc[df['Price'].idxmax()]
most_quantity_product = df.loc[df['Quantity'].idxmax()]

plt.bar(df['Product'], df['Price'], color='lightblue')
plt.title('Product Prices')
plt.xlabel('Products')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()

# Output results
print("=== Inventory Summary ===")
print("\nTotal Value for Each Product:")
print(df[['Product', 'Total Value']])

print(f"\nMost Expensive Product: {most_expensive_product['Product']} with price {most_expensive_product['Price']}")
print(f"Most Available Product: {most_quantity_product['Product']} with quantity {most_quantity_product['Quantity']}")
```

```python
import numpy as np
import pandas as pd

stock1_data = np.array([
    [100, 110, 95],
    [102, 112, 96],
    [104, 115, 98],
    [107, 117, 100],
    [108, 119, 102]
])

stock2_data = np.array([
    [98, 108, 93],
    [101, 111, 95],
    [103, 113, 97],
    [106, 116, 99],
    [107, 118, 101]
])

columns = ['Open', 'High', 'Low']
days = [f'Day {i+1}' for i in range(5)]

df_stock1 = pd.DataFrame(stock1_data, columns=columns, index=days)
df_stock2 = pd.DataFrame(stock2_data, columns=columns, index=days)

sum_matrix = df_stock1 + df_stock2
diff_matrix = df_stock1 - df_stock2

# Display results
print(" Stock 1 Data:")
print(df_stock1, "\n")

print(" Stock 2 Data:")
print(df_stock2, "\n")

print("+ Element-wise Sum (Combined Price Values):")
print(sum_matrix, "\n")

print("- Element-wise Difference (Price Spread):")
print(diff_matrix)
```
